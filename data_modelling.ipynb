DATA CLEANING: (removing duplicates (two rows of the same player), and players who average less than 15 mins per game) (also removing rank row)

```python
df = pd.read_csv('nba_stats333.csv')
cleaned1 = df.drop_duplicates(subset=['FULL NAME'], keep='first')
#print(cleaned1.head(10))
cleaned2 = cleaned1[cleaned1['MPG'] >22]
cleaned3 = cleaned2.drop(columns='RANK')
print(cleaned3.head(20))
```

```python
cleaned3.to_csv('cleaned.csv')
```

```python
print(cleaned3['DRTG'].mean())
```

```python
scoring_=[]
passing_=[]
efficiency_=[]
steals_blocks_=[]
rbg_=[]
rtg_=[]
ovr_scores=[]
names=[]
positions=[]
ages=[]
salaries=[]

#
for i in range(len(cleaned3)):
    playerRow = cleaned3.iloc[i]
    scoring=0
    passing=0
    efficiency=0
    rtg=0
    rbg=0
    steals_blocks=0
    scoringDone=False
    passingDone=False
    efficiencyDone=False
    rtgDone=False
    rbgDone=False
    s_bDone=False
    score=0
    salary=0
    #scoring
    if scoringDone==False:
        if playerRow['PPG']>=25:
            scoring=10
            scoringDone=True
        if (playerRow['PPG']>=22) & (scoringDone==False):
            scoring=9
            scoringDone=True
        if (playerRow['PPG']>=18) & (scoringDone==False):
            scoring=8
            scoringDone=True
        if (playerRow['PPG']>=15) & (scoringDone==False):
            scoring=7
            scoringDone=True
        if (playerRow['PPG']>=10) & (scoringDone==False):
            scoring=6
            scoringDone=True
        if scoringDone==False:
            scoring=5
            scoringDone=True
        scoring_.append(scoring)
    #passing 
    if passingDone==False:
        if (playerRow['APG']>=10):
            passing=10
            passingDone=True
        if (playerRow['APG']>=8) & (passingDone==False):
            passing=9
            passingDone=True
        if (playerRow['APG']>=7) & (passingDone==False):
            passing=8
            passingDone=True
        if (playerRow['APG']>=5) & (passingDone==False):
            passing=7
            passingDone=True
        if (playerRow['APG']>=4) & (passingDone==False):
            passing=6
            passingDone=True
        if passingDone==False:
            passing=5
            passingDone=True
        passing_.append(passing)
    #efficiency 
    if efficiencyDone==False:
        if (playerRow['3P%']>=.4) | (playerRow['eFG%']>=.58):
            efficiency=10
            efficiencyDone=True
        if ((playerRow['3P%']>=.35) | (playerRow['eFG%']>=.55)) & (efficiencyDone==False):
            efficiency=9
            efficiencyDone=True
        if ((playerRow['3P%']>=.32) | (playerRow['eFG%']>=.5)) & (efficiencyDone==False):
            efficiency=8
            efficiencyDone=True
        if ((playerRow['3P%']>=.3) | (playerRow['eFG%']>=.45)) & (efficiencyDone==False):
            efficiency=7
            efficiencyDone=True
        if ((playerRow['3P%']>=.28) | (playerRow['eFG%']>=.42)) & (efficiencyDone==False):
            efficiency=6
            efficiencyDone=True
        if efficiencyDone==False:
            efficiency=5
            efficiencyDone=True
        efficiency_.append(efficiency)
    #rtg 
    if rtgDone==False:
        if (playerRow['ORTG']>=115) | (playerRow['DRTG']>=111):
            rtg=10
            rtgDone=True
        if ((playerRow['ORTG']>=110) | (playerRow['DRTG']>=108)) & (rtgDone==False):
            rtg=9
            rtgDone=True
        if ((playerRow['ORTG']>=108) | (playerRow['DRTG']>=105)) & (rtgDone==False):
            rtg=8
            rtgDone=True
        if ((playerRow['ORTG']>=105) | (playerRow['DRTG']>=102)) & (rtgDone==False):
            rtg=7
            rtgDone=True
        if ((playerRow['ORTG']>=102) | (playerRow['DRTG']>=100)) & (rtgDone==False):
            rtg=6
            rtgDone=True
        if rtgDone==False:
            rtg=5
            rtgDone=True
        rtg_.append(rtg)
    #rbg
    if rbgDone==False:
        if (playerRow['RPG']>=10):
            rbg=10
            rbgDone=True
        if (playerRow['RPG']>=8) & (rbgDone==False):
            rbg=9
            rbgDone=True
        if (playerRow['RPG']>=6) & (rbgDone==False):
            rbg=8
            rbgDone=True
        if (playerRow['RPG']>=5) & (rbgDone==False):
            rbg=7
            rbgDone=True
        if (playerRow['RPG']>=4) & (rbgDone==False):
            rbg=6
            rbgDone=True
        if rbgDone==False:
            rbg=5
            rbgDone=True
        rbg_.append(rbg)
    #stl_blocks
    if s_bDone==False:
        if (playerRow['SPG']>=1) | (playerRow['BPG']>=1):
            steals_blocks=10
            s_bDone=True
        if ((playerRow['SPG']>=.9) | (playerRow['BPG']>=.9)) & (s_bDone==False):
            steals_blocks=9
            s_bDone=True
        if ((playerRow['SPG']>=.7) | (playerRow['BPG']>=.7)) & (s_bDone==False):
            steals_blocks=8
            s_bDone=True
        if ((playerRow['SPG']>=.55) | (playerRow['BPG']>=.55)) & (s_bDone==False):
            steals_blocks=7
            s_bDone=True
        if ((playerRow['SPG']>=.45) | (playerRow['BPG']>=.45)) & (s_bDone==False):
            steals_blocks=6
            s_bDone=True
        if s_bDone==False:
            steals_blocks=5
            s_bDone=True
        steals_blocks_.append(steals_blocks)
    if (playerRow['POS']=='C') | (playerRow['POS']=='F') | (playerRow['POS']=='C-F') | (playerRow['POS']=='F-C'):
        score=(((scoring_[i])*1.15)+((passing_[i])*1.15)+rbg_[i]+rtg_[i]+efficiency_[i]+steals_blocks_[i])/6
    if (playerRow['POS']=='G') | (playerRow['POS']=='G-F') | (playerRow['POS']=='F-G'):
        score=(scoring_[i]+passing_[i]+((rbg_[i])*1.3)+rtg_[i]+efficiency_[i]+steals_blocks_[i])/6
    if (score>=8.6) & (playerRow['AGE']>=23):
        salary=5000
    if (score>=7.8) & (score<8.6) & (playerRow['AGE']>=23):
        salary=3500
    if (score>=7) & (score<7.8) &  (playerRow['AGE']>=23):
        salary=2000
    if (score<7) | (playerRow['AGE']<23):
        salary=1000
    ovr_scores.append(score)
    names.append(playerRow['FULL NAME'])
    ages.append(playerRow['AGE'])
    positions.append(playerRow['POS'])
    salaries.append(salary)

```

```python
data= {'Name': names,
       'Position': positions,
       'Age': ages,
       'Scoring_grade': scoring_,
       'Passing_grade': passing_,
       'Rebounding_grade': rbg_,
       'Rating': rtg_,
       'Efficiency': efficiency_,
       'Turnover_grade': steals_blocks_,
       'Overall': ovr_scores,
       'Salary': salaries,}

fantasyTable = pd.DataFrame(data)

fantasyTable.to_csv('fanTable.csv')
#fantasyDF = pd.dataframe
```
